```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Model
from tensorflow.keras import layers, Input

# 1. Load and preprocess the MNIST dataset
(x_train, _), (x_test, _) = mnist.load_data()

# Normalize the images to the range [0, 1] and flatten them
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.
x_train = x_train.reshape((len(x_train), 28 * 28))
x_test = x_test.reshape((len(x_test), 28 * 28))

# 2. Define the autoencoder architecture
input_dim = x_train.shape[1]  # 784
encoding_dim = 64  # Size of the encoded representations

# Input placeholder
input_img = Input(shape=(input_dim,))

# "encoded" is the encoded representation of the input
encoded = layers.Dense(encoding_dim, activation='relu')(input_img)

# "decoded" is the lossy reconstruction of the input
decoded = layers.Dense(input_dim, activation='sigmoid')(encoded)

# This model maps an input to its reconstruction
autoencoder = Model(input_img, decoded)

# 3. Compile the model
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

# 4. Train the autoencoder
autoencoder.fit(x_train, x_train,
                epochs=50,
                batch_size=256,
                shuffle=True,
                validation_data=(x_test, x_test))

# 5. Use the trained autoencoder to reconstruct images
decoded_imgs = autoencoder.predict(x_test)

# 6. Visualize the original and reconstructed images
n = 10  # Number of digits to display
plt.figure(figsize=(20, 4))
for i in range(n):
    # Display original
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(x_test[i].reshape(28, 28), cmap='gray')
    plt.title("Original")
    plt.axis('off')

    # Display reconstruction
    ax = plt.subplot(2, n, i + 1 + n)
    plt.imshow(decoded_imgs[i].reshape(28, 28), cmap='gray')
    plt.title("Reconstructed")
    plt.axis('off')
plt.show()
```

```python
# =========================
# 1. Import Required Libraries
# =========================
import os
import shutil
import random
import urllib.request
import tarfile
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.applications import VGG16
from tensorflow.keras.optimizers import Adam

import torch
from transformers import BlipProcessor, BlipForConditionalGeneration

# =========================
# 2. Set Constants & Seeds
# =========================
seed_value = 42
random.seed(seed_value)
np.random.seed(seed_value)
tf.random.set_seed(seed_value)

batch_size = 32
n_epochs = 10
img_rows, img_cols = 224, 224
input_shape = (img_rows, img_cols, 3)

# =========================
# 3. Download and Extract Dataset
# =========================
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/ZjXM4RKxlBK9__ZjHBLl5A/aircraft-damage-dataset-v1.tar"
tar_filename = "aircraft_damage_dataset_v1.tar"
extracted_folder = "aircraft_damage_dataset_v1"

# Download tar file
urllib.request.urlretrieve(url, tar_filename)
print(f"Downloaded {tar_filename}. Extraction will begin now.")

# Remove existing folder
if os.path.exists(extracted_folder):
    shutil.rmtree(extracted_folder)
    print(f"Removed the existing folder: {extracted_folder}")

# Extract
with tarfile.open(tar_filename, "r") as tar_ref:
    tar_ref.extractall()
print("Dataset extracted successfully.")

# =========================
# 4. Setup Directories
# =========================
train_dir = os.path.join(extracted_folder, 'train')
valid_dir = os.path.join(extracted_folder, 'valid')
test_dir  = os.path.join(extracted_folder, 'test')

# =========================
# 5. Create Data Generators
# =========================
train_datagen = ImageDataGenerator(rescale=1./255)
valid_datagen = ImageDataGenerator(rescale=1./255)
test_datagen  = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_rows, img_cols),
    batch_size=batch_size,
    class_mode='binary',
    shuffle=True,
    seed=seed_value
)

valid_generator = valid_datagen.flow_from_directory(
    valid_dir,
    target_size=(img_rows, img_cols),
    batch_size=batch_size,
    class_mode='binary',
    shuffle=False
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(img_rows, img_cols),
    batch_size=batch_size,
    class_mode='binary',
    shuffle=False
)

# =========================
# 6. Build Transfer Learning Model (VGG16)
# =========================
base_model = VGG16(weights='imagenet', include_top=False, input_shape=input_shape)

# Freeze all base layers initially
for layer in base_model.layers:
    layer.trainable = False

# Add custom top layers
x = Flatten()(base_model.output)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=output)

# Compile model
model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train model
history = model.fit(
    train_generator,
    epochs=n_epochs,
    validation_data=valid_generator
)

# =========================
# 7. Plot Loss Curves
# =========================
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Loss Over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()

# =========================
# 8. Evaluate on Test Set
# =========================
test_loss, test_acc = model.evaluate(test_generator)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")

# =========================
# 9. Visualize Prediction
# =========================
def plot_prediction_image(test_generator, model, index_to_plot=0):
    test_images, test_labels = next(test_generator)
    predictions = model.predict(test_images)
    predicted_classes = (predictions > 0.5).astype(int).flatten()

    class_names = {v: k for k, v in test_generator.class_indices.items()}
    img = test_images[index_to_plot]
    true_label = int(test_labels[index_to_plot])
    predicted_label = int(predicted_classes[index_to_plot])

    plt.imshow(img)
    plt.title(f"True: {class_names[true_label]}\nPred: {class_names[predicted_label]}")
    plt.axis('off')
    plt.show()

# Test example
plot_prediction_image(test_generator, model, index_to_plot=3)

# =========================
# 10. BLIP Captioning & Summary
# =========================
# Load BLIP model
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
blip_model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# BLIP text generation function
def generate_text(image_path_tensor, task_tensor):
    def process_image(image_path, task):
        image_path_str = image_path.numpy().decode("utf-8")
        task_str = task.numpy().decode("utf-8")
        image = Image.open(image_path_str).convert("RGB")

        prompt = "This is a picture of" if task_str == "caption" else "This is a detailed photo showing"
        inputs = processor(images=image, text=prompt, return_tensors="pt")
        output = blip_model.generate(**inputs)
        return processor.decode(output[0], skip_special_tokens=True)

    return tf.py_function(process_image, [image_path_tensor, task_tensor], tf.string)

# Example image path
image_path = tf.constant("aircraft_damage_dataset_v1/test/dent/149_22_JPG_jpg.rf.4899cbb6f4aad9588fa3811bb886c34d.jpg")

# Generate and print results
caption = generate_text(image_path, tf.constant("caption"))
summary = generate_text(image_path, tf.constant("summary"))

print("Generated Caption:", caption.numpy().decode("utf-8"))
print("Generated Summary:", summary.numpy().decode("utf-8"))

# Display the image
img = plt.imread(image_path.numpy().decode("utf-8"))
plt.imshow(img)
plt.axis('off')
plt.title("Example Image")
plt.show()
```

```python
import glob

# =========================
# 10. Test BLIP on 10 Images
# =========================

# Get 10 random image paths from the test set
image_paths = glob.glob("aircraft_damage_dataset_v1/test/*/*.jpg")  # Adjust for file type if needed
random.shuffle(image_paths)
sampled_images = image_paths[:10]

# Loop through and generate caption & summary for each
for idx, path in enumerate(sampled_images):
    print(f"\n--- Image {idx+1} ---")

    image_path = tf.constant(path)

    # Generate caption
    caption = generate_text(image_path, tf.constant("caption"))
    caption_text = caption.numpy().decode("utf-8")

    # Generate summary
    summary = generate_text(image_path, tf.constant("summary"))
    summary_text = summary.numpy().decode("utf-8")

    # Display results
    print("Caption:", caption_text)
    print("Summary:", summary_text)

    # Show image
    img = plt.imread(path)
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"Image {idx+1}")
    plt.show()
```

```python
# Ensure test_generator is not exhausted
test_generator.reset()

# Get one batch of images and labels (will contain up to batch_size images)
test_images, test_labels = next(test_generator)

# Get predictions from the trained CNN model
predictions = model.predict(test_images)
print(f'################################### {predictions}')
predicted_classes = (predictions > 0.5).astype(int).flatten()
print(f'################################### {predicted_classes}')


# Get class label names from the generator
class_names = {v: k for k, v in test_generator.class_indices.items()}
print(f'################################### {class_names}')



# Show 10 results
for i in range(10):
    # Image array to disk temporarily to use with BLIP
    image_array = (test_images[i] * 255).astype(np.uint8)  # undo normalization
    print(f'###################################ARR {image_array}')

    image_pil = Image.fromarray(image_array)
    temp_path = f"temp_image_{i}.jpg"
    image_pil.save(temp_path)

    # Generate caption and summary using BLIP
    tf_image_path = tf.constant(temp_path)
    caption = generate_text(tf_image_path, tf.constant("caption"))
    summary = generate_text(tf_image_path, tf.constant("summary"))

    # Decode predictions
    true_label = int(test_labels[i])
    pred_label = int(predicted_classes[i])

    # Show results
    print(f"\n--- Image {i+1} ---")
    print(f"True Label     : {class_names[true_label]}")
    print(f"Predicted Label: {class_names[pred_label]}")
    print("Caption        :", caption.numpy().decode("utf-8"))
    print("Summary        :", summary.numpy().decode("utf-8"))

    # Plot image
    plt.imshow(test_images[i])
    plt.title(f"True: {class_names[true_label]} | Pred: {class_names[pred_label]}")
    plt.axis('off')
    plt.show()

    # Optionally remove temp file
    os.remove(temp_path)
```
